```python
import matplotlib.pyplot as plt
from enmspring.spring import Spring
from enmspring.k_b0_util import get_df_by_filter_bp
```

### Part 0: Initialization

```python
rootfolder = '/home/yizaochen/codes/dna_rna/fluctmatch_sequence'
cutoff = 4.7
host = 'ctct_21mer'
type_na = 'bdna+bdna'
n_bp = 21
spring_obj = Spring(rootfolder, host, type_na, n_bp)
```

### Part 1: Read DataFrame

```python
df = spring_obj.read_k_b0_pairtype_df_given_cutoff(cutoff)
```

### Part 2: Get HB DataFrame

```python
df_hb = get_df_by_filter_bp(df, 'hb')
```

```python
df_hb
```

```python
k_dict = dict()
for resid in range(1, 22):
    mask = (df_hb['Strand_i'] == 'STRAND1') & (df_hb['Resid_i'] == resid)
    df_temp = df_hb[mask]
    k_dict[resid] = df_temp['k'].tolist()
```

```python
k_list = list()
for resid in range(1, 22):
    k_list.append(k_dict[resid])
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(16,6))

ax.boxplot(k_list)

plt.show()
```
